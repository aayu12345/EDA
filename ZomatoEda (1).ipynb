```python
final_df.Country.value_counts()
# we just want to know that how many records for countries are there
```

```python
country_names=final_df.Country.value_counts().index
# this will give us the  all countries name so that we can store in any variable
```

```python
country_val=final_df.Country.value_counts().values
#this will give you the numpayarray for which you can plot pie chart
```

```python

plt.pie(country_val,labels=country_names)
plt.show()
# this will give you some data which is not understandable
#so we are going to perform it for top 3 or 5 countries
```

```python

plt.pie(country_val[:3],labels=country_names[:3])
plt.show()
```

```python
# so now i want to add some percentage so we use autopct here
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
plt.show()
```
